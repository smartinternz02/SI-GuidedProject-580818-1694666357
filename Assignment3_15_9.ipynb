# Outlier Detection

```python
#outlier detection for Age only
out=[]
def Zscore_outlier(df):
    m = np.mean(df)
    sd = np.std(df)
    for i in df:
        z = (i-m)/sd
        if np.abs(z) > 3:
            out.append(i)
    print("Outliers:",out)
Zscore_outlier(df['Age'])
```

```python
#deleting outlier in Age as they are small in number
sns.boxplot(df['Age'])
plt.title("Box Plot Aefore Removing Outlier")
plt.show()

#removing outlier
iqr = 1.5 * (np.percentile(df['Age'], 75) - np.percentile(df['Age'], 25))
df.drop(df[df['Age'] > (iqr + np.percentile(df['Age'], 75))].index, inplace=True)
df.drop(df[df['Age'] < (np.percentile(df['Age'], 25) - iqr)].index, inplace=True)

#showing boxplot after removing age outlier
sns.boxplot(df['Age'])
plt.title("Box Plot After Removing Outlier")
plt.show()



```
