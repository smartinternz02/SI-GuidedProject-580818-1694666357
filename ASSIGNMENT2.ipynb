## **Scatter Plot**

```python
sns.scatterplot(x='speeding',y='total',data=df)
# higher speed lead to more accident
```

```python
sns.scatterplot(x='alcohol',y='total',data=df)
```

```python
sns.scatterplot(x='not_distracted',y='total',data=df)
#people who are less distracted has
```

```python
 sns.scatterplot(x='ins_premium',y='ins_losses',data=df)
# people who paid less insurance got less loss
```

# **Bar Plot**

```python
sns.barplot(x='speeding',y='total',data=df,ci=None)
```

```python
sns.barplot(x="speeding",y="alcohol",data=df.head(5),ci=None)
```

```python
sns.barplot(y='total',x='abbrev',hue='alcohol',data=df.head(10))
# DC has least accident with least alcohol ie 1.5932000000002
```
