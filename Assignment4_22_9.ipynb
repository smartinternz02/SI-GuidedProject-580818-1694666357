# Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
columns = ['Attrition','OverTime','BusinessTravel', 'Department','Education',
                       'EducationField','JobSatisfaction','EnvironmentSatisfaction','WorkLifeBalance',
                       'StockOptionLevel','Gender', 'PerformanceRating', 'JobInvolvement',
                       'JobLevel', 'JobRole', 'MaritalStatus','RelationshipSatisfaction','Over18']
```

```python
df1=df.copy()
```

```python
df1[columns]=df1[columns].apply(le.fit_transform)
```

```python
df1.head()
```

```python
df1.info()
```

# Spliting independent and dependent variable


```python
df1.corr()
```

```python
x = df1.drop(columns='Attrition')
```

```python
y = df1['Attrition']
```

```python
x.head()
```

```python
y.head()
```

# Feature Scaling 

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ms=MinMaxScaler()
ms
```

```python
x_scaled=pd.DataFrame(ms.fit_transform(x),columns=x.columns)
```

```python
x_scaled    
```

# Splitting Data into Train and test Split

```python

from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y,test_size=0.2,random_state=0)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

# LOGISTIC REGRESSION

```python
from sklearn.linear_model import LogisticRegression
```

```python
model_logistic=LogisticRegression()
model_logistic
```

```python
model_logistic.fit(x_train,y_train)
```

```python
pred_logistic=model_logistic.predict(x_test)
```

```python
pred_logistic
```

```python
y_test
```

# DECISION TREE CLASSIFIER

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model_decision_tree=DecisionTreeClassifier()
model_decision_tree
```

```python
model_decision_tree.fit(x_train,y_train)
```

```python
pred_decision_tree=model_decision_tree.predict(x_test)
```

```python
pred_decision_tree
```

```python
y_test
```

# Evaluation of Classificaton model

## For Logistic Regression Model

```python
#Accuracy score
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,roc_auc_score,roc_curve
```

```python
accuracy_score_logistic=accuracy_score(y_test,pred_logistic)
accuracy_score_logistic
```

```python
confusion_matrix_logistic=confusion_matrix(y_test,pred_logistic)
confusion_matrix_logistic
```

```python
crosstab_logistic=pd.crosstab(y_test,pred_logistic)
crosstab_logistic
```

```python
print(classification_report(y_test,pred_logistic))
```

```python
probability_logistic=model_logistic.predict_proba(x_test)[:,1]
```

```python
probability_logistic
```

```python
# Roc curve
fpr_logistic,tpr_logistic,threshsholds_logistic = roc_curve(y_test,probability_logistic)
```

```python
plt.plot(fpr_logistic,tpr_logistic)
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC CURVE FOR LOGISTIC REGRESSION MODEL')
plt.show()
```

## For Decision Tree Model
